```python
from datetime import date
import httpx

today = date.today()
data = {
    "query": f"publication-date={today.strftime("%Y%m%d")}",
    "fields": [
        "publication-date",
        "notice-title",
        "procedure-type",
        "contract-nature",
        # TODO fix country
        # "country",
        "tender-value",
        "tender-value-cur",
        "classification-cpv",
        "organisation-contact-point-tenderer",
        "document-url-lot"
    ],
    "page": 1,
    "limit": 10,
    "scope": "ACTIVE",
    "checkQuerySyntax": False,
    "paginationMode": "PAGE_NUMBER"
}

r = httpx.post('https://api.ted.europa.eu/v3/notices/search', json=data)
r.json()
```
